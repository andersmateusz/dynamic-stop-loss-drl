```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

## Generate sample data

```python
np.random.seed(42)
feature_1 = np.random.randint(0, 100, size=100)
feature_2 = np.random.randint(1000, 5000, size=100)
feature_1 = np.append(feature_1, 500)
feature_2 = np.append(feature_2, 10000)
data = np.column_stack((feature_1, feature_2))
```

## Min-max vs Z-score normalization

```python
# Min-Max Normalization
min_max_scaler = MinMaxScaler()
data_min_max = min_max_scaler.fit_transform(data)

# Z-Score Normalization
z_score_scaler = StandardScaler()
data_z_score = z_score_scaler.fit_transform(data)

# Plotting the results
plt.figure(figsize=(14, 6))

plt.subplot(1, 3, 1)
plt.scatter(data[:, 0], data[:, 1], color='black')
plt.title('Before normalization')
plt.xlabel('X')
plt.ylabel('Y')
plt.subplot(1, 3, 2)
plt.scatter(data_min_max[:, 0], data_min_max[:, 1], color='black')
plt.title('Min-Max normalization')
plt.xlabel('X')
plt.ylabel('Y')

plt.subplot(1, 3, 3)
plt.scatter(data_z_score[:, 0], data_z_score[:, 1], color='black')
plt.title('Z-Score normalization')
plt.xlabel('X')
plt.ylabel('Y')

plt.tight_layout()

plt.savefig('../charts/normalization.eps', format='eps')
plt.show()
```
