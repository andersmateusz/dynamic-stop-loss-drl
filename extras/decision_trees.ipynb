```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Sample panel data

```python
data = {
    'income': [int(np.random.random()*10000.0+3000) for _ in range(10)],
    'age_group': [np.random.choice(['Young','Middle-Aged','Senior']) for _ in range(10)],
    'purchase': [np.random.choice([0,1]) for _ in range(10)]
}

df = pd.DataFrame(data)
print(df.to_latex(index=False))
df
```

## Derive rules from the data

```python
# Convert 'Age_Group' to numerical categories
df['age_group'] = df['age_group'].map({'Young': 0, 'Middle-Aged': 1, 'Senior': 2})
X = df[['income', 'age_group']]
y = df['purchase']
# Train a decision tree classifier
clf = DecisionTreeClassifier(random_state=0, max_depth=3)
clf.fit(X, y)
plt.figure(figsize=(14, 16))
plot_tree(clf, feature_names=['Income', 'Age group'], class_names=['No', 'Yes'], filled=True)
plt.savefig('../charts/decision_trees_classifier.eps', format='eps')
plt.show()
```
